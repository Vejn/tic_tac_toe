```python
game_list = [[0,1,2],[3,4,5],[6,7,8]]

def display_game(game_list):
    print("Here is the current list: ")
    print(game_list[0])
    print(game_list[1])
    print(game_list[2])
    
def position_choice():
    
    choice = "wrong"
    
    while choice not in ['0','1','2','3','4','5','6','7','8']:
        
        choice = input("pick a free position from the list: ")
        
        if choice not in ['0','1','2','3','4','5','6','7','8']:
            print("Sorry, invalid choice!")
    
    if choice == "0":
        posa = 0
        posb = 0
    elif choice == "1":
        posa = 0
        posb = 1
    elif choice == "2":
        posa = 0
        posb = 2
    elif choice == "3":
        posa = 1
        posb = 0
    elif choice == "4":
        posa = 1
        posb = 1
    elif choice == "5":
        posa = 1
        posb = 2
    elif choice == "6":
        posa = 2
        posb = 0
    elif choice == "7":
        posa = 2
        posb = 1
    elif choice == "8":
        posa = 2
        posb = 2
    
    position = (posa,posb)
    
    return position

def replacement_choice(game_list,position):
    user_placement = "Wrong"
    while user_placement not in ["X","x","O","o"]:
        user_placement = input("Type 'X' or 'O' to place at position: ")
        if user_placement in ["X","x","O","o"]:
            game_list[position[0]][position[1]] = user_placement
        
    return game_list

def gameon_choice():
    
    choice = "wrong"
    
    while choice not in ["Y","y","N","n"]:
        
        choice = input("Keep playing?Y/N: ")
        
        if choice not in ["Y","y","N","n"]:
            print("Sorry, invalid choice!")
    if choice in["Y","y"]:
        return True
    else:
        return False
    
game_on = True
game_list = [[0,1,2],[3,4,5],[6,7,8]]

while game_on:
    display_game(game_list)
    
    position = position_choice()
    
    game_list = replacement_choice(game_list,position)
    
    display_game(game_list)
    
    game_on = gameon_choice()
```
